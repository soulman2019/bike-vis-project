Now what about the speed (km/h)?

```python
plt.figure()
data.speed.plot(kind='hist', bins=120, edgecolor='black', facecolor='white')
plt.xlim(0, 22)
plt.title('Speed distribution (km/h)')
plt.show()
```

The speed appears to be normally distributed except that there is a significant outlier around 0. This may come from users who don't ride their bike directly to the destination but takes a long break on the road. This can also be seen from the long tail of the duration distribution.

We can use this observation about the outlier in speed to create a new feature: break. We infer that a user has taken a break if they have a speed of less than 1 km/h. You can walk significantly faster than this and so it seems safe to conclude that the trip included a stop/break.

```python
plt.figure()
data['break'].value_counts(sort=False).plot(kind='bar')
plt.title('Break distribution')
plt.show()
```

The distribution of breaks look suspiciously like the trip type distribution. Let's make a confusion matrix of the two features overlap.

```python
trip_type_break = np.array([
    [
        len(data[(data['break'] == 'No') & (data['trip_type'] == 'Single')]),
        len(data[(data['break'] == 'No') & (data['trip_type'] == 'Return')]),
        len(data[(data['break'] == 'No') & (data['trip_type'] == 'Unknown')]),
    ],
    [
        len(data[(data['break'] == 'Yes') & (data['trip_type'] == 'Single')]),
        len(data[(data['break'] == 'Yes') & (data['trip_type'] == 'Return')]),
        len(data[(data['break'] == 'Yes') & (data['trip_type'] == 'Unknown')]),
    ],
    [
        len(data[(data['break'] == 'Unknown') & (data['trip_type'] == 'Single')]),
        len(data[(data['break'] == 'Unknown') & (data['trip_type'] == 'Return')]),
        len(data[(data['break'] == 'Unknown') & (data['trip_type'] == 'Unknown')]),
    ],
])
trip_type_break = trip_type_break / len(data)
plt.figure()
plt.imshow(trip_type_break)
plt.xticks([0, 1, 2], ['No', 'Yes', 'Unknown'])
plt.yticks([0, 1, 2], ['Single', 'Return', 'Unknown'])
for x in range(3):
    for y in range(3):
        value = trip_type_break[x, y]
        color = 'black' if value > 0.5 else 'white'
        plt.text(x, y, f'{value:0.3f}', horizontalalignment="center", color=color)
plt.xlabel('Break')
plt.ylabel('Trip type')
plt.show()
```

We see that about 2/3 of the trips with breaks are return trips.

```python
breaks = data[data['break'] == 'Yes']
percentage = len(breaks[breaks['trip_type'] == 'Return']) / len(breaks)
print(f'Fraction of trips with breaks that are return trips: {percentage:0.2%}')
```

This confirms what we saw in the confusion matrix.

The distribution is not as spiked as it is for duration but we do see a clear morning spike which returns in the late afternoon - similar to the distribution of speed in a day. Speaking of speed, let's look at how it distributes between genders in a day.

```python
data_male = data[data.member_gender == 'Male']
data_female = data[data.member_gender == 'Female']
```

```python
plt.figure()
speed_pr_hour_male = data_male.groupby('start_hour').speed.mean()
speed_pr_hour_female = data_female.groupby('start_hour').speed.mean()
speed_pr_hour_male.plot(label='Male')
speed_pr_hour_female.plot(label='Female')
plt.xlabel('Start hour')
plt.ylabel('Avg. speed (km/h)')
plt.title('Avg. speed in a day')
plt.legend()
plt.ylim(0, 12)
plt.show()
```

This corresponds to what we would expect since men, biologically speaking, tend to be physically stronger than women. We should also check if the distance travelled is different between the genders.

```python
plt.figure()
distance_pr_hour_male = data_male.groupby('start_hour').distance_vincenty_km.mean()
distance_pr_hour_female = data_female.groupby('start_hour').distance_vincenty_km.mean()
distance_pr_hour_male.plot(label='Male')
distance_pr_hour_female.plot(label='Female')
plt.xlabel('Start hour')
plt.ylabel('Avg. distance (km)')
plt.title('Avg. distance in a day')
plt.legend()
plt.ylim(0, 3)
plt.show()
```

The distance travelled appears to be roughly the same but there is an interesting thing to notice here: Women travel more in the morning than men do. Perhaps it might be that it is because women tend to rent more bikes in the morning? (Which could be because women tend to get up earlier than men?)

```python
plt.figure()
count_pr_hour_male = data_male.groupby('start_hour').bike_id.count()
count_pr_hour_female = data_female.groupby('start_hour').bike_id.count()
count_pr_hour_male.plot(label='Male')
count_pr_hour_female.plot(label='Female')
plt.xlabel('Start hour')
plt.ylabel('#Bike rentals')
plt.title('#Bike rentals in a day')
plt.legend()
plt.show()
```

This graph doesn't help a lot since there are a lot more male than female users. We need to normalize to be able to compare.

```python
plt.figure()
count_pr_hour_male = data_male.groupby('start_hour').bike_id.count() / len(data_male)
count_pr_hour_female = data_female.groupby('start_hour').bike_id.count() / len(data_female)
count_pr_hour_male.plot(label='Male')
count_pr_hour_female.plot(label='Female')
plt.xlabel('Start hour')
plt.ylabel('#Normalized bike rentals')
plt.title('#Normalized bike rentals in a day')
plt.legend()
plt.show()
```

This shows that there are no significant differences in when men and women rent a bike (except for a small spike in the morning). Perhaps there are differences in when they stop using a bike?

```python
plt.figure()
count_pr_hour_male = data_male.groupby('end_hour').bike_id.count() / len(data_male)
count_pr_hour_female = data_female.groupby('end_hour').bike_id.count() / len(data_female)
count_pr_hour_male.plot(label='Male')
count_pr_hour_female.plot(label='Female')
plt.xlabel('End hour')
plt.ylabel('#Normalized bike rentals')
plt.title('#Normalized bike rentals in a day')
plt.legend()
plt.show()
```

Nope - the same as for start hour.
